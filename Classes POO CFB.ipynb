```python
class Carro:
  velmax=0
  ligado=False
  cor=""
  def __init__(self, v,l,c):
    self.velmax=v
    self.ligado=l
    self.cor=c
  def mostrar(self):
    print(f"Velocidade Max:{self.velmax} ")
    estado= "sim" if self.ligado else "Nao"
    print(f"Ligado:{estado} ")
    print(f"Cor:{self.cor} ")
  def ligar(self):
    self.ligado = True
  def desligar(self):
    self.ligado = False
  def andar(self):
    if self.ligado:
      print("Ligado e Andando")
    else:
      print("Desligado, Ligue o carro.")

c1=Carro(200, False, "vermelho")
c2=Carro(100, False, "Amarelo")
c3=Carro(130, False, "Preto")
c1.ligar()
c1.mostrar() 
c1.desligar()
c1.mostrar()
```
