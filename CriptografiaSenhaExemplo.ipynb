```python
#Cifra de Cezar

#A Variavel alfabeto recebe recebe uma lista contento 'str' de A a Z.
alfabeto=['a','b','c','d','e','f','g','h','i','j','l','m','n','o','p','q','r','s','t','u','v','x','z']
#Variavel r recebe 3. r sera o controle do incremento. "Ex A criptografado seria C"
r=3
#Usuario entra com dados de 4 digitos.
senha=input("digite senha 4 digitos: ")
#.index() serve para retornar um indice de um objeto na lista. Ex: "print(alfabeto.index('a'))" = 0 
#"print(alfabeto.index('b')+r)" = 4
#d0 a d3. Variavel(s) recebe o caracter digitado pelo usuario
#O .index() retorna a posição do indice do caracter digitado da lista alfabeto.
d0=(alfabeto.index(senha[0])+r)%len(alfabeto)
d1=(alfabeto.index(senha[1])+r)%len(alfabeto)
d2=(alfabeto.index(senha[2])+r)%len(alfabeto)
d3=(alfabeto.index(senha[3])+r)%len(alfabeto)

#menssagem exibida criptografada.
print("Menssagem criptografada:",alfabeto[d0],alfabeto[d1],alfabeto[d2],alfabeto[d3])


#ic=(i+r)%len(alfabeto)
#id=(i-r)%len(alfabeto)
```
