```python
#Programa completo. Jogo PAR ou Impar
import random
print("Jogo da Adivinhação")

print("Como jogar: Escolha um numero de 1 a 5")

print("Depois escolha PAR(p) ou IMPAR(i)")

print("O numero o qual você escolheu sera somado ao numero que o computador escolheu.")

chance=3
score=float()

while chance>-1:

  meu_numero=int(input("Escolha um numero (1 a 5) "))

  minha_escolha=input("Escolha PAR(p) ou IMPAR(i)")

  pc=random.randint(1,5)

  result=int(pc+meu_numero)
  print(f"Suas Chances: {chance}")
  if result%2==0:
    if minha_escolha.upper()=="P":
      score+=1
      chance=chance
      print(f"Você GANHOU o resultado deu PAR :{result} / Chances: {chance} / Score: {score} ")
      
    else:
      chance=chance-1
      print(f"Você PERDEU o resultado deu PAR :{result} / Chances: {chance} / Score: {score} ")
      
  else:
    if minha_escolha.upper()=="I":
      score+=1
      chance=chance
      print(f"Você GANHOU o resultado deu IMPAR :{result} / Chances: {chance} / Score: {score} ")
      
    else:
      chance=chance-1
      print(f"Você PERDEU o resultado deu IMPAR :{result} / Chances: {chance} / Score: {score} ")
      
  if chance==0:
    print(f"FIM DE JOGO / Score: {score} ")
    break
```
