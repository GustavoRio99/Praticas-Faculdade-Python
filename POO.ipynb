```python
#Aluno é o tipo do OBJETO
class Aluno:
  #Metodo construtor da classe
  def __init__(self, pNome, pIdade, pN1=0, pN2=0, pN3=0): 
    #Atributos
    self.nome= pNome
    self.idade= pIdade
    self.n1= pN1
    self.n2= pN2
    self.n3= pN3
  #Metodos
  def media(self):
    resultado=(self.n1+self.n2+self.n3)/3
    return resultado
#A1 é o objeto que recebe Classe Alunos e seus parametros
a1=Aluno(a1.nome, a1.idade, a1.n1, a1.n2, a1.n3)
a1.nome=input("Nome Aluno:")
a1.idade=input("idade:")
a1.n1=float(input("Digite sua Primeira nota: "))
a1.n2=float(input("Digite sua Segunda nota: "))
a1.n3=float(input("Digite sua Terceira nota: "))
```

```python
#Objeto a1 acessa o atributo nome e Idade da Classe Alunos e imprime.
print("Nome:",a1.nome,",",a1.idade,"anos")
#Objeto a1 acessa o Metodo media(). Acessa e Calcula os valores do Objeto a1.n1, a1.n2. a1.n3
print("Media:",a1.media())
```

```python
class Pessoa:
  def __init__(self, pNome, pIdade):
    self.nome = pNome
    self.idade = pIdade
  def Andar(self):
    print("andando")
#PESSOA FISICA HERDANDO CPF CLASSE PESSOA
class PessoaFisica(Pessoa):
    def __init__(self, pNome, pIdade, pCpf):
      Pessoa.__init__(self, pNome, pIdade)
      self.cpf = pCpf


pf=PessoaFisica("Gustavo", 30, 45465111)
print(pf.nome, pf.idade, pf.cpf)
pf.Andar()
```

```python
class Pessoa:
  def __init__(self, pNome, pIdade):
    self.nome = pNome
    self.idade = pIdade
  def Andar(self):
    print("andando")
#PESSOA FISICA HERDANDO CPF CLASSE PESSOA
class PessoaFisica(Pessoa):
    def __init__(self, pNome, pIdade, pCpf):
      Pessoa.__init__(self, pNome, pIdade)
      self.cpf = pCpf
class PessoaJuridica(Pessoa):
  def __init__(self, pNome, pIdade, pCnpj):
    super().__init__(pNome, pIdade)
    self.cnpj = pCnpj

pj=PessoaJuridica("Gustavo", 30, 15465846854)
pf=PessoaFisica("bompreço", 30, 45465111)
print(pf.nome, pf.idade, pf.cpf)
print(pj.nome, pj.idade, pj.cnpj)
```
