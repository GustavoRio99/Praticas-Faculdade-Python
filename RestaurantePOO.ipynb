```python
class Restaurante:
  def __init__(self, pNome, pCozinha):
    self.nome=pNome
    self.cozinha=pCozinha
  def desc(self):
    print("Nome: ",self.nome,"Cozinha: ",self.cozinha)
  def abrir(self):
    print("RESTAURANTE: ",self.nome,"ABERTO CARALHO!!!") 

obRest=Restaurante("LAMOLE","VEGANA ME ENGANA")
obRest.desc()
obRest.abrir()
```
