```python
from random import randint

print("Jogo par ou IMPAR")
meu_numero=int(input("Digite qualquer numero entre 1 e 10"))
minha_escolha=input("Escolha PAR ou IMPAR: ")
pc=randint(1,10)
result=meu_numero+pc

if result % 2==0:
  if minha_escolha.upper()=='P':
    print("Você ganhou a soma dos numeros deu PAR")
  else:
    print("Você perdeu a soma dos numeros deu PAR")
else:
  if minha_escolha.upper()=='I':
    print("Você ganhou a soma dos numeros é IMPAR")
  else:
    print("Você perdeu a soma dos numeros é IMPAR")

print(f"O numero que você digitou: {meu_numero}")
print(f"O numero que o PC escolheu: {pc}")
print(f"O numeros somados: {result}")

```
